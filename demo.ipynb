# Loading annotated dataframe

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from numpy.polynomial.polynomial import polyfit
```

```python
df = pd.read_csv('annotated_df.csv')
```

```python
df
```

# Vizualizing

```python
df['score'] = df['score'].astype(float)
df['rel_score'] = df['rel_score'].astype(float)
df['ratio_score'] = df['ratio_score'].astype(float)
```

```python
right = df[df['correctness'] == 'yes']
wrong = df[df['correctness'] == 'no']
```

```python
right_pos = right['position'].to_numpy()
right_scor = right['score'].to_numpy()
right_rel_scor = right['rel_score'].to_numpy()
right_rat_scor = right['ratio_score'].to_numpy()

wrong_pos = wrong['position'].to_numpy()
wrong_scor = wrong['score'].to_numpy()
wrong_rel_scor = wrong['rel_score'].to_numpy()
wrong_rat_scor = wrong['ratio_score'].to_numpy()
```

```python
# Absolute score vs. position

plt.scatter(right_scor, right_pos, color='blue', label='Correct')
plt.scatter(wrong_scor, wrong_pos, color='red', label='Incorrect')
plt.legend(loc='lower left')
plt.xlabel('score')
plt.ylabel('position')
plt.savefig("absolute_score_vs_position.png", dpi=150)
```

```python
# Relative score vs. position

plt.scatter(right_rel_scor, right_pos, color='blue', label='Correct')
plt.scatter(wrong_rel_scor, wrong_pos, color='red', label='Incorrect')
plt.legend(loc='lower right')
plt.xlabel('rel_score')
plt.ylabel('position')
plt.savefig("relative_score_vs_position.png", dpi=150)
```

```python
# Ratio score vs. position

plt.scatter(right_rat_scor, right_pos, color='blue', label='Correct')
plt.scatter(wrong_rat_scor, wrong_pos, color='red', label='Incorrect')
plt.legend(loc='lower left')
plt.xlabel('ratio_score')
plt.ylabel('position')
plt.savefig("ratio_score_vs_position.png", dpi=150)
```

```python
# Absolute score vs. rel_score

plt.scatter(right_scor, right_rel_scor, color='blue', label='Correct')
plt.scatter(wrong_scor, wrong_rel_scor, color='red', label='Incorrect')
plt.legend(loc='upper right')
plt.xlabel('absolute_score')
plt.ylabel('rel_score')
plt.savefig("absolute_score_vs_rel_score.png", dpi=150)
```

```python
# Absolute score vs. ratio score

plt.scatter(right_scor, right_rat_scor, color='blue', label='Correct')
plt.scatter(wrong_scor, wrong_rat_scor, color='red', label='Incorrect')
plt.legend(loc='lower right')
plt.xlabel('absolute_score')
plt.ylabel('ratio_score')
plt.savefig("absolute_score_vs_ratio_score.png", dpi=150)
```

```python
# Relative score vs. ratio score

plt.scatter(right_rel_scor, right_rat_scor, color='blue', label='Correct')
plt.scatter(wrong_rel_scor, wrong_rat_scor, color='red', label='Incorrect')
plt.legend(loc='upper right')
plt.xlabel('relative_score')
plt.ylabel('ratio_score')
plt.savefig("relative_score_vs_ratio_score.png", dpi=150)
```
